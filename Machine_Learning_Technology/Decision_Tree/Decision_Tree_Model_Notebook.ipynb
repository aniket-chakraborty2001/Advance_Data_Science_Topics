```python
# Importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Read the 'Employee_Salaries.csv' data
data = pd.read_csv('/content/drive/MyDrive/Datasets/Employee_Salaries.csv')
data
```

```python
# Separating the input and output features
# Inputs as X and output as y
X = data.drop(['salary_more_then_100k'], axis = 1)
y = data['salary_more_then_100k']
```

```python
# ML algorithms doesn't understand labels
# Need to label encode the comapny, job and degree column
# Need to create three label ecoder objects for three different columns
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
# Using the fit_transform method to convert them into integer form
X['company'] = le_company.fit_transform(X['company'])
X['job'] = le_company.fit_transform(X['job'])
X['degree'] = le_company.fit_transform(X['degree'])
X
```

```python
# Define the Decision Tree Classifier object
model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
```

```python
# Fit the X and y data in the model
model.fit(X,y)
```
