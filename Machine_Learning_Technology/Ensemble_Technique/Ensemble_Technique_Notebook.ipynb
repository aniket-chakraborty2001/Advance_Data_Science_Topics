```python
# Importing requied packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
```

```python
# Reading the 'Pima_Indian_Diabetes_Data.csv' file
data = pd.read_csv('/content/drive/MyDrive/Datasets/Pima_Indian_Diabetes_Data.csv')
data.head()
```

```python
# Check for any missing values
data.isnull().sum()
```

```python
# Obtain basic statistical summary from the data set
data.describe()
```

```python
# Check for any imbalance in the 'Outcome' column
data['Outcome'].value_counts()

# Imbalance is there, but it is not very much
```

```python
# Creating the input(X) and output(y) variable
X = data.drop(['Outcome'], axis = 1)
y = data['Outcome']
```

```python
# Use standard scaler to scale the features of X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Get the First three entries
X_scaled[:3]
```

```python
# Creating the training and testing sample
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify  = y, random_state = 20)
print('Shape of X_train is:', X_train.shape)
print('Shape of X_test is:', X_test.shape)
print('Shape of y_train is:', y_train.shape)
print('Shape of y_test is:', y_test.shape)
```

```python
# Use Decision Tree Classifie Model: Define model object, train the model, test the model - prediction
# Use cross validation for this purpose with 5 flods
scores = cross_val_score(DecisionTreeClassifier(), X, y, cv = 5)
scores
```

```python
# Get the accuracy of the decision tree model
scores.mean()

# The model is 71% accurate
```

```python
# Now create a BaggingClassifier with DecisionTree Classifier, consider 100 models, with 80% samples
# oob_score is set to True (oob_score: Out of bag score, the sample that did't appear in small samples, can be used as test sample)
# oob_score is the accuracy of the model
bag_model = BaggingClassifier(
    estimator = DecisionTreeClassifier(),
    n_estimators = 100,
    max_samples = 0.8,
    oob_score = True,
    random_state = 0
)
bag_model.fit(X_train, y_train)
bag_model.oob_score_
```

```python
# Runing the bag_model with cross validation method
scores = cross_val_score(bag_model, X, y, cv = 5)
scores.mean()

# See, the accuracy increases from 71% to 75%
```

```python
# Use the RandomForestClassifier model with the Cross validation method
scores = cross_val_score(RandomForestClassifier(), X, y, cv = 5)
scores.mean()

# The accuracy of the model increases further
```
