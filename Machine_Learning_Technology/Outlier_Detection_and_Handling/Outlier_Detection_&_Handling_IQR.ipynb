```python
# Importing required packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the 'Person_Heights_Data.csv' file
data = pd.read_csv('/content/drive/MyDrive/Datasets/Person_Heights_Data.csv')
data.head()
```

```python
# Get some summary statistics
data.describe()
```

```python
# Creating some imaginary data points
new_data = pd.DataFrame({'name':['Aniket', 'Sayan','Riju', 'Arka', 'Tuhin'],
            'height':[5.27,3.41,10.21,32.25,1.25]})
new_data
```

```python
# Adding those new data points in original data frame
data = pd.concat([data, new_data], axis = 'rows')
data.head()
```

```python
# Get summary statistics of the data dataframe
data.describe()
```

```python
# Calculate the IQR for the height column
q1 = data['height'].quantile(0.25)
q3 = data['height'].quantile(0.75)
IQR = q3-q1
IQR
```

```python
# Determine the uuper limit and lower limit for height column
low_lim = q1 - 1.5 * IQR
up_lim = q3 + 1.5 * IQR
print('The Lower and Upper limits are:', low_lim, up_lim)
```

```python
# Get the shape of the data dataframe
data.shape
```

```python
# Anything outside the range (low_lim, up_lim) is considered as outlier
# Ommit the outliers from the data
clean_data = data[(data['height'] > low_lim) & (data['height'] < up_lim)]
clean_data.shape
```
