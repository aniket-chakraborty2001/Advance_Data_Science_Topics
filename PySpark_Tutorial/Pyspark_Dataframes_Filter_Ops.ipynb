```python
# Get the path of the 'PySpark_test1.csv' file as file_path
file_path = '/content/drive/MyDrive/Datasets/PySpark_test1.csv'
```

```python
# Importing Pyspark
import pyspark
```

```python
# Importing a SparkSession and creating a spark context object as spark with 'FilterOps' app name
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('FilterOps').getOrCreate()
```

```python
# Check the details of the Spark context object
spark
```

```python
# Read the 'PySpark_test1.csv' file as df_org with first row as header and inferSchema set to True
df_org = spark.read.csv(file_path, header = True, inferSchema = True)
df_org.show()
```

```python
# Get the all informations for the people having saraly less than or equal to 20000
df_org.filter('Salary <= 20000').show()

# This can also be written as the following code
df_org.filter(df_org['Salary'] <= 20000).show()
```

```python
# Get the 'Name' and 'age' of the people having salary less than or equal to 20000
df_org.filter('Salary <= 20000').select(['Name','age']).show()
```

```python
# Get all the informations for people having salary in the range [15000,20000]
df_org.filter((df_org['Salary'] <= 20000) | (df_org['Salary'] >= 15000)).show() # The operaor '|' means and
```

```python
# Get all information of people having salary greater than 20000
df_org.filter(df_org['Salary'] > 20000).show()

# This can also be thought as: people not having salary less than or equal to 20000
df_org.filter(~(df_org['Salary']<=20000)).show()
```
